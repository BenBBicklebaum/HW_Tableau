```python
import pandas as pd
import numpy as np
```

```python
csv = '202001-citibike-tripdata.csv'
df =  pd.read_csv(csv)
df.head()
```

```python
df["tripmin"] = (df["tripduration"] / 60).round(2)

df.head()
```

```python
df['Trip'] = ""
df.head()
```

```python
df = df.sort_values(['bikeid', 'starttime'], ascending=[True, True])
df.reset_index(drop=True, inplace=True)
df.head()
```

```python
df['Trip'] = df.groupby('bikeid').cumcount() + 1
df.head(100)
```

```python
df["gender"].replace({0: "NA", 1: "M", 2: "F"}, inplace=True)
df.head()
```

```python
df.to_csv("bikeride.csv", index=False)
```
